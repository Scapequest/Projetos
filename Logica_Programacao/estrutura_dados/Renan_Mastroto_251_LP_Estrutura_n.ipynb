### 3. Elabore uma estrutura para representar um produto (código, nome, preço). **Crie uma função** para cadastrar 5 produtos. **Crie outra função** para aplicar 10% de aumento no preço do produto e apresente, por meio **de outra função**, todos os dados do produtos cadastrados, após o aumento. Construa uma função para cada opção do menu a seguir:
Menu do Sistema
1. Cadastrar
2. Reajustar
3. Visualizar
4. Sair

Qual opção deseja?

```python
import random

class Produto:
  cod = 0
  nome = ""
  preco = 0.0
  preco_n = 0.0

def menu():
  print('\nMenu de Opções')
  print('1 - Cadastrar')
  print('2 - Reajustar')
  print('3 - Visualizar')
  print('4 - Sair')
  return int(input('Digite a opção desejada: '))

def cadastrar(produtos):
  for i in range(2):
    prod = Produto()
    prod.cod = i
    prod.nome = input("Digite o nome do produto: ")
    prod.preco = random.randint(1, 100)
    produtos.append(prod)
  return produtos

def reajustar(produtos):
  for i in range(len(produtos)):
    prod = produtos[i]
    prod.preco_n = prod.preco * 1.1
  return produtos

def visualizar(produtos):
  if len(produtos) > 0:
    print("\n|{0:^10}|{1:^25}|{2:^20}|{3:^20}|".format("Cód","Nome","R$ Antigo","R$ Novo"))
    for i in range(len(produtos)):
      prod = produtos[i]
      print("|{0:^10}|{1:^25}|{2:^20}|{3:^20}|".format(prod.cod,prod.nome,f"R$ {prod.preco:.2f}", f"R$ {prod.preco_n:.2f}"))
  else:
    print("\nNão há produtos cadastrados")

def main():
  produtos = []
  while True:
    opcao = menu()
    if opcao == 1:
      produtos = cadastrar(produtos)
    elif opcao == 2:
      produtos = reajustar(produtos)
    elif opcao == 3:
      visualizar(produtos)
    elif opcao == 4:
      break
    else:
      print("Opção Inválida")

main()
```

Levando em conta que esta escola possui no máximo 500 alunos, como você faria para estruturar estas informações num sistema de gerenciamento para a escola? Implemente utilizando estrutura.
```python
class TipoAluno:
  nome = ''
  dt_nasc = ''
  fone = 0
  endereco = ''
  serie = 0

def menu():
  print('\nMenu de Opções')
  print('1 - Cadastrar')
  print('2 - Consultar')
  print('3 - Visualizar')
  print('4 - Sair')
  return int(input('Digite a opção desejada: '))

def cadastro(alunos):
  aluno = TipoAluno()
  aluno.nome = input("Digite o nome do aluno: ")
  aluno.dt_nasc = input("Digite a data de nascimento: ")
  aluno.fone = int(input("Digite o telefone para contato: "))
  aluno.endereco = input("Digite o endereço do aluno: ")
  aluno.serie = int(input("Digite a série do aluno: "))
  alunos.append(aluno)
  return alunos

def consultar(alunos):
  nome = input("\nDigite o nome para consultar: ")
  print('\n{0:^5}|{1:^20}|{2:^15}|{3:^15}|{4:^30}|{5:^7}'.format("Cód", "Nome", "Nascimento", "Telefone", "Endereço", "Série"))
  for i in range(len(alunos)):
    aluno = alunos[i]
    if nome.upper() in aluno.nome.upper():
      print('{0:^5}|{1:^20}|{2:^15}|{3:^15}|{4:^30}|{5:^7}'.format(i,aluno.nome,aluno.dt_nasc,aluno.fone,aluno.endereco,aluno.serie))

def visualizar(alunos):
  if len(alunos) > 0:
    print('\n{0:^5}|{1:^20}|{2:^15}|{3:^15}|{4:^30}|{5:^7}'.format("Cód", "Nome", "Nascimento", "Telefone", "Endereço", "Série"))
    for i in range(len(alunos)):
      aluno = alunos[i]
      print('{0:^5}|{1:^20}|{2:^15}|{3:^15}|{4:^30}|{5:^7}'.format(i,aluno.nome,aluno.dt_nasc,aluno.fone,aluno.endereco,aluno.serie))
  else:
    print("\nNão há alunos cadastrados")

def main():
  alunos = []
  while True:
    opcao = menu()
    if opcao == 1:
      alunos = cadastro(alunos)
    elif opcao == 2:
      consultar(alunos)
    elif opcao == 3:
      visualizar(alunos)
    elif opcao == 4:
      break
    else:
      print("Opção Inválida")

main()
```

### 5. Faça um programa que realize o cadastro de contas bancarias com as seguintes informações: numero da conta, nome do titular e saldo. O banco permitirá o cadastramento de 15 contas. **Crie uma função para cada opção do menu** a seguir. Utilize a estrutura na passagem de parâmetro:
Menu de opções:
1. Cadastrar contas
2. Visualizar todas as contas
3. Consultar por nome
4. Alterar nome e/ou saldo de um número de conta
5. Excluir a conta com menor saldo
6. Sair

Observação:

* No item de menu 4. Alterar nome e/ou saldo de um número de conta, entenda que apenas pode ser alterado o nome e o saldo depois que você pesquisar pelo número da conta.
* No item 5 do menu, encontre o menor saldo dentre todos cadastrados, **depois exclua esta ÚNICA conta.**. O assunto de excluir algo de um vetor foi dado na disciplina de Algoritmo.



```python
import random
contas_teste = list(range(1000,2000))

class TipoConta:
  numero_conta = 0
  titular = ''
  saldo = 0.0

def menu():
  print('\nMenu de Opções')
  print('1 - Cadastrar')
  print('2 - Visualizar')
  print('3 - Consultar')
  print('4 - Alterar Nome/Saldo')
  print('5 - Excluir Menor Conta')
  print('6 - Sair')
  return int(input('Digite a opção desejada: '))

def cadastro(contas):
  conta = TipoConta()
  conta.titular = input("Digite o nome do titular da conta: ")
  conta.numero_conta = contas_teste.pop(random.randint(0,len(contas_teste)-1))
  conta.saldo = float(input("Digite o valor do saldo: R$ "))
  contas.append(conta)
  return contas

def consultar(contas):
  nome = input("\nDigite o nome para consultar: ")
  print('\n|{0:^17}|{1:^17}|{2:^15}|'.format("Número da Conta", "Nome do Titular", "Saldo"))
  for i in range(len(contas)):
    conta = contas[i]
    if nome.upper() in conta.titular.upper():
      print('|{0:^17}|{1:^17}|{2:^15}|'.format(conta.numero_conta, conta.titular, f"R$ {conta.saldo:.2f}"))

def visualizar(contas):
  if len(contas) > 0:
    print('\n|{0:^17}|{1:^17}|{2:^15}|'.format("Número da Conta", "Nome do Titular", "Saldo"))
    for i in range(len(contas)):
      conta = contas[i]
      print('|{0:^17}|{1:^17}|{2:^15}|'.format(conta.numero_conta, conta.titular, f"R$ {conta.saldo:.2f}"))
  else:
    print("\nNão há contas cadastrados")

def alterar(contas):
  num = int(input("\nDigite o número da conta que deseja alterar: "))
  seletor = len(contas)+2
  for i in range(len(contas)):
    if contas[i].numero_conta == num:
      seletor = i
  if seletor == len(contas)+2:
    print("Conta não encontrada")
  else:
    while True:
      opcao_alt = int(input("Selecione o que deseja alterar:\n1 - Nome\n2 - Saldo\n3 - Sair\n"))
      if opcao_alt == 1:
        contas[seletor].titular = input("Digite o novo nome: ")
      elif opcao_alt == 2:
        contas[seletor].saldo = float(input("Digite o novo saldo: R$ "))
      elif opcao_alt == 3:
        break
      else:
        print("Opção Inválida")
  return contas

def remover(contas):
  menor = contas[0].saldo
  menor_n = 0
  for i in range(len(contas)):
    if contas[i].saldo < menor:
      menor = contas[i].saldo
      menor_n = i
  opcao_rem = input(f"\nRemovendo conta {contas[menor_n].numero_conta}\nDeseja continuar? (S/N)")
  if opcao_rem.upper() == "S":
    contas.pop(menor_n)
    print("Conta removida")
  elif opcao_rem.upper() == "N":
    print("Cancelando")
  return contas

def main():
  contas = []
  while True:
    opcao = menu()
    if opcao == 1:
      contas = cadastro(contas)
    elif opcao == 2:
      visualizar(contas)
    elif opcao == 3:
      consultar(contas)
    elif opcao == 4:
      contas = alterar(contas)
    elif opcao == 5:
      contas = remover(contas)
    elif opcao == 6:
      break
    else:
      print("Opção Inválida")

main()
```

### 7. Elabore uma estrutura para representar um Produto (código, nome, data_fabricacao, data_validade, preço). Para o membro data_fabricacao e data_validade, deve-se criar outra estrutura Data (dia, mês, ano). **Utilize aninhamento de estruturas** para resolver este desenvolvimento. Construa uma função para cada opçao do menu a seguir:

Menu de opções:
1. Cadastrar produtos
2. Visualizar todos os dados
3. Sair

```python
class TipoData:
    dia = 0
    mes = 0
    ano = 0

class TipoProduto:
    codigo = 0
    nome = ''
    data_fabricacao = TipoData()
    data_validade = TipoData()
    preco = 0.0

def menu():
  print('\nMenu de Opções')
  print('1 - Cadastrar Produtos')
  print('2 - Visualizar Todos')
  print('3 - Sair')
  return int(input('Digite a opção desejada: '))

def cadastro(produtos, codigos):
  while True:
    prod = TipoProduto()
    prod.data_fabricacao = TipoData()
    prod.data_validade = TipoData()
    prod.codigo = codigos.pop(0)
    prod.nome = input("\nDigite o nome do produto: ")
    prod.preco = float(input("Digite o valor do produto: R$ "))
    prod.data_fabricacao.dia = int(input("Digite o dia de fabricação: "))
    prod.data_fabricacao.mes = int(input("Digite o mês de fabricação: "))
    prod.data_fabricacao.ano = int(input("Digite o ano de fabricação: "))
    prod.data_validade.dia = int(input("Digite o dia de validade: "))
    prod.data_validade.mes = int(input("Digite o mês de validade: "))
    prod.data_validade.ano = int(input("Digite o ano de validade: "))
    produtos.append(prod)
    if input("Deseja cadastrar mais? (S/N) ").upper() != "S":
      break
  return produtos, codigos

def maior(prod):
  m_nome = len(prod[0].nome)
  for i in range(len(prod)):
    if len(prod[i].nome) > m_nome:
      m_nome = len(prod[i].nome)
  return m_nome

def visualizar(produtos):
  if len(produtos) > 0:
    m_nome = maior(produtos)
    print(f'\n|{"CÓDIGO":^8}|{"PRODUTO":^{m_nome + 4}}|{"PREÇO":^11}|{"FABRICAÇÃO":^14}|{"VALIDADE":^14}|')
    for i in range(len(produtos)):
      prod = produtos[i]
      fabricacao = f"{prod.data_fabricacao.dia:02}/{prod.data_fabricacao.mes:02}/{prod.data_fabricacao.ano}"
      validade = f"{prod.data_validade.dia:02}/{prod.data_validade.mes:02}/{prod.data_validade.ano}"
      print(f'|{prod.codigo:^8}|{prod.nome:^{m_nome + 4}}|{f"R$ {prod.preco:.2f}":^11}|{fabricacao:^14}|{validade:^14}|')
  else:
    print("\nNão há produtos cadastrados")

def main():
  produtos = []
  codigos = list(range(1000,2000))
  while True:
    opcao = menu()
    if opcao == 1:
      produtos, codigos = cadastro(produtos, codigos)
    elif opcao == 2:
      visualizar(produtos)
    elif opcao == 3:
      break
    else:
      print("Opção Inválida")

main()
```

```python
import random
import string

class TipoCliente:
  cod_cli = 0
  nome = ''
  fone = ''

class TipoDocumento:
  num_doc = 0
  cod_cli = 0
  dia_venc = 0
  dia_pag = 0
  valor = 0.0
  juros = 0.0

def menu():
  print('\nSistema de Gerenciamento de Documentos e Clientes')
  print('1 - Cadastrar clientes')
  print('2 - Relatório de clientes')
  print('3 - Cadastrar documentos')
  print('4 - Relatório de documentos')
  print('5 - Excluir clientes sem documentos')
  print('6 - Excluir documento por número')
  print('7 - Excluir documento por cliente')
  print('8 - Excluir documento por período')
  print('9 - Alterar dados de cliente')
  print('10 - Quantidade de Documentos por Cliente')
  print('11 - Sair')
  print('12 - Função Teste - Gerar n clientes com documentos aleatóriamente')
  return int(input('Qual opção deseja? '))

def cadastroCliente(clientesV, codigos, doc_dic):
  while True:
    cliente = TipoCliente()
    cliente.nome = input("\nDigite o nome do cliente: ").title()
    cliente.fone = int(input("Digite o telefone do cliente: "))
    doc_dic[codigos[0]] = []
    cliente.cod_cli = codigos.pop(0)
    clientesV.append(cliente)
    if input("Deseja cadastrar mais? (S/N) ").upper() != "S":
      break
  return clientesV, codigos, doc_dic

def maior(clientesV):
  m_nome = len(clientesV[0].nome)
  for i in range(len(clientesV)):
    if len(clientesV[i].nome) > m_nome:
      m_nome = len(clientesV[i].nome)
  return m_nome

def relatorioClientes(clientesV, doc_dic):
  if len(clientesV) > 0:
    m_nome = maior(clientesV)
    print(f'\n|{"CÓDIGO":^8}|{"CLIENTE":^{m_nome + 4}}|{"TELEFONE":^12}|{"QTDE DOCUMENTOS":^19}|')
    for i in range(len(clientesV)):
      cliente = clientesV[i]
      print(f'|{cliente.cod_cli:^8}|{cliente.nome:^{m_nome + 4}}|{cliente.fone:^12}|{len(doc_dic[cliente.cod_cli]):^19}|')
  else:
    print("\nNão há clientes cadastrados")

def cadastrarDocumentos(doc_dic, num_doc):
  chave = int(input("\nDigite o código do cliente: "))
  if chave in doc_dic:
    while True:
      documento = TipoDocumento()
      documento.num_doc = num_doc.pop(0)
      documento.cod_cli = chave
      documento.valor = float(input("Digite o valor do documento: R$ "))
      documento.dia_pag = int(input("Digite o dia do pagamento: "))
      documento.dia_venc = int(input("Digite o dia do vencimento: "))
      if documento.dia_pag > documento.dia_venc:
        documento.juros = documento.valor * 0.05
      doc_dic[chave].append(documento)
      if input("Deseja cadastrar mais? (S/N) ").upper() != "S":
        break
  else:
    print("Código inválido!")
  return doc_dic, num_doc

def relatorioDocumentos(doc_dic):
  print(f'\n|{"Nº DOC":^8}|{"CÓD CLIENTE":^13}|{"VALOR SEM JUROS":^20}|{"VALOR COM JUROS":^20}|{"DIA PAGAMENTO":^15}|{"DIA VENCIMENTO":^16}|')
  for i in doc_dic:
    for j in doc_dic[i]:
      print(f'|{j.num_doc:^8}|{j.cod_cli:^13}|{f"R$ {j.valor:.2f}":^20}|{f"R$ {j.valor+j.juros:.2f}":^20}|{f"{j.dia_pag:02}":^15}|{f"{j.dia_venc:02}":^16}|')

def teste(clientes, doc_dic, codigos, num_doc, quant):
  for i in range(quant):
    cliente = TipoCliente()
    cliente.nome = ''.join(random.choices(string.ascii_letters, k=random.randint(5, 10))).title()
    cliente.fone = 39030000+ random.randint(1, 9999)
    doc_dic[codigos[0]] = []
    if random.randint(0,10) % 2 == 0:
      for j in range(random.randint(1,2)):
        documento = TipoDocumento()
        documento.num_doc = num_doc.pop(0)
        documento.cod_cli = codigos[0]
        documento.valor = random.random() * 1000
        documento.dia_pag = random.randint(1,31)
        documento.dia_venc = random.randint(1,31)
        if documento.dia_pag > documento.dia_venc:
          documento.juros = documento.valor * 0.05
        doc_dic[codigos[0]].append(documento)
    cliente.cod_cli = codigos.pop(0)
    clientes.append(cliente)
  return clientes, doc_dic, codigos, num_doc

def excluirVazios(clientes, doc_dic, codigos):
  for i in doc_dic:
    if len(doc_dic[i]) == 0:
      codigos.insert(0, int(i))
      for indice, cliente in enumerate(clientes):
        if cliente.cod_cli == int(i):
          clientes.pop(indice)
  return clientes, doc_dic, codigos

def excluirDocumento(doc_dic, num_doc):
  num = int(input("\nDigite o número do documento que deseja excluir: "))
  if num in num_doc:
    print("Documento não encontrado!")
  else:
    for i in doc_dic:
      for indice, objeto in enumerate(doc_dic[i]):
        if objeto.num_doc == num:
          num_doc.append(num)
          doc_dic[i].pop(indice)
    print(f"Documento Nº {num} excluído!")
  return doc_dic, num_doc

def excluirDocumentoCliente(doc_dic, codigos, num_doc):
  num = int(input("\nDigite o código do cliente: "))
  if num in codigos:
    print("Cliente não encontrado!")
  else:
    doc_deletados = []
    for j, i in enumerate(doc_dic[num]):
      doc_deletados.append(int(i.num_doc))
    doc_dic[num] = []
    print(doc_deletados)
    num_doc.extend(doc_deletados)
  return doc_dic, num_doc

def excluirDocumentoPeriodo(doc_dic, num_doc):
  periodo_inicio = int(input('\nDigite o início do período:'))
  periodo_fim = int(input('Digite o final do período:'))
  periodo = list(range(periodo_inicio,periodo_fim+1))
  doc_deletados = []
  for i in doc_dic:
    for indice, objeto in enumerate(doc_dic[i]):
      if objeto.dia_venc in periodo:
        doc_deletados.append(objeto.num_doc)
        doc_dic[i].pop(indice)
  print(f"Documentos Excluídos: {len(doc_deletados)}")
  num_doc.extend(doc_deletados)
  return doc_dic, num_doc

def alterarDadosCliente(clientes, codigos):
  while True:
    cliente = int(input("\nDigite o código do cliente para alterar: "))
    if cliente in codigos:
      print("Cliente não cadastrado!")
    else:
      for i in clientes:
        if i.cod_cli == cliente:
          novo_nome = input("Digite o novo nome do cliente (Deixe vazio para manter): ").title()
          novo_fone = int(input("Digite o novo telefone do cliente (Digite 0 para manter): "))
          if len(novo_nome) > 0:
            i.nome = novo_nome
          if novo_fone != 0:
            i.fone = novo_fone
    if input("Desejar alterar outro (S/N)? ").title() != "S":
      break
  return clientes

def relatorioDocumentosClientes(clientesV, doc_dic, codigos):
  cliente = int(input("\nDigite o código do cliente desejado: "))
  if cliente not in codigos:
    for i in clientesV:
      if i.cod_cli == cliente:
        print(f'\n|{"CÓDIGO":^8}|{"CLIENTE":^{len(i.nome)}}|{"TELEFONE":^12}|{"QTDE DOCUMENTOS":^19}|')
        print(f'|{i.cod_cli:^8}|{i.nome:^{len(i.nome)}}|{i.fone:^12}|{len(doc_dic[i.cod_cli]):^19}|')
  else:
    print("\nCliente não cadastrado")

def main():
  clientes = []
  doc_dic = {}
  codigos = list(range(1000,2000))
  num_doc = list(range(200,2200))
  while True:
    codigos.sort()
    num_doc.sort()
    clientes.sort(key=lambda x: x.cod_cli)
    opcao = menu()
    if opcao == 1:
      clientes, codigos, doc_dic = cadastroCliente(clientes, codigos, doc_dic)
    elif opcao == 2:
      relatorioClientes(clientes, doc_dic)
    elif opcao == 3:
      doc_dic, num_doc = cadastrarDocumentos(doc_dic, num_doc)
    elif opcao == 4:
      relatorioDocumentos(doc_dic)
    elif opcao == 5:
      clientes, doc_dic, codigos = excluirVazios(clientes, doc_dic, codigos)
    elif opcao == 6:
      doc_dic, num_doc = excluirDocumento(doc_dic, num_doc)
    elif opcao == 7:
      doc_dic, num_doc = excluirDocumentoCliente(doc_dic, codigos, num_doc)
    elif opcao == 8:
      doc_dic, num_doc = excluirDocumentoPeriodo(doc_dic, num_doc)
    elif opcao == 9:
      clientes = alterarDadosCliente(clientes, codigos)
    elif opcao == 10:
      relatorioDocumentosClientes(clientes, doc_dic, codigos)
    elif opcao == 11:
      break
    elif opcao == 12:
      quant = int(input("Digite a quantidade de clientes para gerar: "))
      clientes, doc_dic, codigos, num_doc = teste(clientes, doc_dic, codigos, num_doc, quant)
    else:
      print("Opção Inválida")

main()
```

####9. Uma empresa prestadora de serviços armazena informações sobre os serviços prestados. Sabe-se que a empresa pode realizar no máximo três serviços diariamente. É de interesse de sua direção manter um histórico mensal (30 dias) sobre os serviços prestados.
A empresa realiza quatro tipos de serviços:

Cada serviço realizado deve ser cadastrado com as seguintes informações: número do serviço, valor do serviço, código do serviço e código do cliente, **numa matriz 30x3 referente a estrutura Servico_prestado**.

Cadastre/digite os quatro tipos de serviços (código e descrição) que a empresa poderá realizar. Para isso, utilize um vetor de quatro posições referente a **estrutura Tipo_servico**.
O programa deverá mostrar o seguinte menu de opções:
1. Cadastrar os tipos de serviços
2. Mostrar todos os tipos de serviço
3. Cadastrar os serviços prestados

**Para a opção 1**: deve-se cadastrar os tipos de serviços oferecidos pela empresa, com código e descrição.

**Para a opção 3**: deve-se considerar que deverão ser cadastrados os serviços prestados ao logo do mês. Em cada dia podem ser cadastrados, no máximo, três serviços prestados.

Utilize uma matriz capaz de armazenar em cada posição todas as informações referentes a um serviço prestado (número, valor, código do serviço, código do cliente). **Cada linha representa um dia do mês, mas a matriz inicia na linha zero, então se o usuário quer serviço para o dia 1, deverá ser gravado na linha zero, se for dia 14, será na linha 13, e assim sucessivamente**. Dessa maneira, considere a matriz com dimensão 30 × 3. DICA, crie a matriz com valores zeros para todos os campos na main().

Poderão existir linhas vazias, pois corresponderão a sábados, domingos ou feriados.

Solicite o dia em que o serviço foi prestado e as demais informações.
Lembre-se de que a empresa só pode prestar os serviços que já tenham sido cadastrados no vetor de tipo de serviços.

Caso o usuário digite um código de tipo de serviço inválido, o programa deverá mostrar uma mensagem de erro.

Quando o usuário tentar cadastrar mais de três serviços prestados em um mesmo dia, também deverá mostrar uma mensagem de erro.

**Para a opção 5**: o programa deverá receber o dia que se deseja consultar e mostrar os respectivos serviços prestados.

**Para a opção 6**: o programa deverá receber o valor mínimo e o valor máximo e mostrar os serviços prestados que estiverem nesse intervalo.

**Para a opção 7**: o programa deverá mostrar todos os serviços prestados, conforme o exemplo a seguir.

                  DIA 01


No do serviço| Valor do serviço R$| Código do serviço| Descrição| Código do cliente
---|---|---|---|---
100| 200,00| 1| Pintura| 1
150| 100,00| 3| Faxina| 5

                  DIA 02

---|---|---|---|---

```python
import random

class TipoServico:
  nome = ''
  cod = 0

class TipoServicoPrestado:
  numero_servico = 0
  cod_servico = 0
  cod_cliente = 0
  valor = 0.0
  nome = ''

def cadastrar_servico(tipos_servicos):
  if len(tipos_servicos) < 4:
    for i in range((4-len(tipos_servicos))):
      servico = TipoServico()
      servico.nome = input(f"\nDigite o nome do serviço: ").title()
      servico.cod = len(tipos_servicos)+1
      tipos_servicos.append(servico)
      print(f"Código do Serviço: {servico.cod} | Nome do Serviço: {servico.nome}")
  else:
    print("Serviços já cadastrados!")
  return tipos_servicos

def cadastrarServicoPrestado(numero_servico, tipos_servicos, servicos_prestados):
  while True:
    dia = int(input("\nDigite o dia do serviço prestado: ")) - 1
    if len(servicos_prestados[dia]) == 3:
      print("Máximo de serviços prestados no dia!")
      continue
    else:
      servico_prestado = TipoServicoPrestado()
      servico_prestado.numero_servico = numero_servico.pop(0)
      while True:
        cod = int(input("Digite o código do serviço: "))
        if cod > 0 and cod < 5:
          servico_prestado.cod_servico = cod
          break
        else:
          print("Código de Serviço errado!")
      servico_prestado.nome = tipos_servicos[cod-1].nome
      servico_prestado.valor = float(input("Digite o valor do serviço: R$ "))
      servico_prestado.cod_cliente = int(input("Digite o código do cliente: "))
      servicos_prestados[dia].append(servico_prestado)
    if input("Deseja cadastras mais (S/N)? ").title() != "S":
      break
  return numero_servico, tipos_servicos, servicos_prestados

def listar_servicos(tipos_servicos):
  print(f'\n|{"CÓDIGO":^8}|{"SERVIÇO":^15}|')
  for i in range(len(tipos_servicos)):
    print(f'|{tipos_servicos[i].cod:^8}|{tipos_servicos[i].nome:^15}|')

def listar_servicos_prestados(servicos_prestados, tipos_servicos):
  contagem = [0] * len(tipos_servicos)
  total = [0.0] * len(tipos_servicos)
  for i in range(len(servicos_prestados)):
    for j in servicos_prestados[i]:
      indice = int(j.cod_servico) - 1
      if 0 <= indice < len(tipos_servicos):
        contagem[indice] += 1
        total[indice] += float(j.valor)
  total_valor = sum(total)
  if total_valor > 0:
    print(f'\n|{"SERVIÇO":<15}|{"QUANTIDADE":^10}|{"VALOR TOTAL":^15}|')
    for i in range(len(contagem)):
      print(f'|{tipos_servicos[i].nome:<15}|{contagem[i]:^10}|R$ {total[i]:^12.2f}|')
    print(f'\n{"TOTAL GERAL ARRECADADO:":<27}R$ {total_valor:^12.2f}')
  else:
    print("\nNenhum serviço prestado cadastrado.")

def listar_servicos_por_data(servicos_prestados, tipos_servicos):
  dia = int(input("\nDigite o dia desejado: "))
  if len(servicos_prestados[dia-1]) > 0:
    print(f'|{"Nº SERVIÇO":^14}|{"VALOR":^11}|{"CÓDIGO SERVIÇO":^16}|{"DESCRIÇÃO":^15}|{"CÓDIGO CLIENTE":^16}|')
    for i, j in enumerate(servicos_prestados[dia-1]):
      print(f'|{j.numero_servico:^14}|{f"R$ {j.valor}":^11}|{j.cod_servico:^16}|{tipos_servicos[j.cod_servico-1].nome:^15}|{j.cod_cliente:^16}|')
  else:
    print("Sem serviços cadastrados no dia!")

def listar_servicos_por_valor(servicos_prestados, tipos_servicos):
  inicial = int(input("\nDigite o valor inicial: R$"))
  final = int(input("Digite o valor final: R$"))
  print(f'|{"Nº SERVIÇO":^14}|{"VALOR":^11}|{"CÓDIGO SERVIÇO":^16}|{"DESCRIÇÃO":^15}|{"CÓDIGO CLIENTE":^16}|')
  for i in servicos_prestados:
    for j in i:
      if j.valor >= inicial and j.valor <= final:
        print(f'|{j.numero_servico:^14}|{f"R$ {j.valor}":^11}|{j.cod_servico:^16}|{tipos_servicos[j.cod_servico-1].nome:^15}|{j.cod_cliente:^16}|')

def gerar_relatorio_geral(servicos_prestados, tipos_servicos):
  for i in range(len(servicos_prestados)):
    if len(servicos_prestados[i]) > 0:
      print('\n{0:^72}'.format(f'========DIA {i+1:02}========'))
      print(f'|{"Nº SERVIÇO":^14}|{"VALOR":^11}|{"CÓDIGO SERVIÇO":^16}|{"DESCRIÇÃO":^15}|{"CÓDIGO CLIENTE":^16}|')
      for j in servicos_prestados[i]:
        print(f'|{j.numero_servico:^14}|{f"R$ {j.valor}":^11}|{j.cod_servico:^16}|{tipos_servicos[j.cod_servico-1].nome:^15}|{j.cod_cliente:^16}|')
    else:
      continue

def teste(tipos_servicos):
  teste_servicos = ["Faxina", "Manutenção", "Jardinagem", "Pintura"]
  for i in range(int(input())):
    servico = TipoServico()
    servico.cod = len(tipos_servicos)+1
    servico.nome = teste_servicos.pop(random.randrange(len(teste_servicos)))
    tipos_servicos.append(servico)
  return tipos_servicos

# ↑↓ DUAS FUNÇÕES DE TESTE QUE CADASTRAM DADOS ALEATÓRIAMENTE E FACILITA O TESTE DAS OUTRAS FUNÇÕES

def teste_dois(numero_servico, tipos_servicos, servicos_prestados):
  selecao = [random.choice(list(range(4))) for i in range(len(servicos_prestados))]
  for i in range(len(servicos_prestados)):
    if selecao[i] == 0:
      continue
    else:
      for j in range(selecao[i]):
        servico_prestado = TipoServicoPrestado()
        servico_prestado.numero_servico = numero_servico.pop(0)
        cod = random.randint(1,4)
        servico_prestado.cod_servico = cod
        servico_prestado.nome = tipos_servicos[cod-1]
        servico_prestado.valor = round(random.uniform(1,9) * 100)
        servico_prestado.cod_cliente = random.randint(1,100)
        servicos_prestados[i].append(servico_prestado)
  return numero_servico, tipos_servicos, servicos_prestados

def exibir_menu():
  print("\nOpções:")
  print("1. Cadastrar tipos de serviços")
  print("2. Mostrar todos os tipos de serviços")
  print("3. Cadastrar serviços prestados")
  print("4. Mostrar todos os serviços prestados")
  print("5. Mostrar serviços prestados em determinado dia")
  print("6. Mostrar serviços prestados dentro de um intervalo de valor")
  print("7. Mostrar um relatório geral (separado por dia), que exiba, inclusive, a descrição do tipo do serviço")
  print("8. Sair")
  return input("Escolha uma opção: ")

def main():
  numeros_servico = list(range(1,100))
  tipos_servicos = []
  servicos_prestados = [[] for j in range(30)]
  while True:
    opcao = exibir_menu()
    if opcao == '1':
      tipos_servicos = cadastrar_servico(tipos_servicos)
    elif opcao == '2':
      listar_servicos(tipos_servicos)
    elif opcao == '3':
      numeros_servico, tipos_servicos, servicos_prestados = cadastrarServicoPrestado(numeros_servico, tipos_servicos, servicos_prestados)
    elif opcao == '4':
      listar_servicos_prestados(servicos_prestados, tipos_servicos)
    elif opcao == '5':
      listar_servicos_por_data(servicos_prestados, tipos_servicos)
    elif opcao == '6':
      listar_servicos_por_valor(servicos_prestados, tipos_servicos)
    elif opcao == '7':
      gerar_relatorio_geral(servicos_prestados, tipos_servicos)
    elif opcao == '8':
      print("Saindo do sistema.")
      break
    # OPÇÃO 9 E 10 CHAMAM AS FUNÇÕES DE TESTE
    elif opcao == '9':
      tipos_servicos = teste(tipos_servicos)
    elif opcao == '10':
      numeros_servico, tipos_servicos, servicos_prestados = teste_dois(numeros_servico, tipos_servicos, servicos_prestados)
    else:
      print("Opção inválida. Tente novamente.")

main()
```
